# Complex numbers

Complex numbers in Python are generally of the form `a + bj` where `a` and `b` are numbers and `j` is the imaginary unit. The letter $j$ is a common alternative to $i$ for the imaginary unit; $i$ is more common in mathematics (and quantum computing) while $j$ is more common in engineering. For consistency, I will continue to use $i$ in all contexts except verbatim Python code.  

```python
z = 1 + 1j
x = 2 + 2j

print(x+z)
```

The $a + bi$ form of complex numbers is sometimes called "rectangular form" because the components can be visualized as two orthogonal dimensions in a cartesian coordinate system. The form $r\cdot e^{i\theta}$ is known as the "exponential" or "polar" form of complex numbers. Any complex number can be expressed in either form, but in quantum computing we use the polar form a lot because of its nice representation of points around the unit circle. Note that since we are often interested in the unit circle $r=1$ so we will often see the polar form without $r$ written. 

In order to convert between rectangular and polar form in Python, we can use the `cmath` library. 

```python
import cmath

z = 3 + 4j
r, theta = cmath.polar(z)

print(f"Magnitude (r): {r}")          # Output: 5.0
print(f"Phase (theta): {round(theta, 2)}")   # Output: 0.93 (radians)
```

The other way is possible too. For this, we'll also import `math` in order to convert from degrees to radians. For the most part, library functions use radians, but often degrees are more intuitive for people. In this class, you will definitely want to become as comfortable as possible with radians. 

```python
import math

# Define polar coordinates
r = 5
theta_degrees = 53.13

# Convert degrees to radians (standard library functions expect radians)
theta_radians = math.radians(theta_degrees)

# Convert to rectangular form
z_rect = cmath.rect(r, theta_radians)
rounded_z_rect = complex(round(z_rect.real, 2), round(z_rect.imag, 2))

print(f"Rectangular form: {rounded_z_rect}")
# Output: Rectangular form: (3.000030113840673+3.9999774148151776j)
```

### Graphing complex numbers

We can use Python's `matplotlib` to visualize complex numbers on the complex plane. 

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(z.real, z.imag, color='red')

# Add lines for axes
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Labels and grid
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.grid(True)

plt.show()
```

We can also graph the same number in polar form

```python
from fractions import Fraction
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')

ax.plot(theta, r, 'ro')  # 'ro' for a red dot
# Label the point
ax.annotate(f" {z}",
    xy=(theta, r),
    xytext=(-20, 8), # These values can be used to adjust the relative location of the point label
    textcoords='offset points', ha='left', va='bottom')

# Converts x-tick labels from degrees to radians. The default is to display as degrees
# Get the x-tick positions (returns in radians)
label_positions = ax.get_xticks()
# Convert to a list since we want to change the type of the elements
labels = list(label_positions)
# Make the labels display as fractions of pi
labels = [rf'$\frac{{{Fraction(label/np.pi).numerator}}}{{{Fraction(label/np.pi).denominator}}}$'+"π" for label in labels]
ax.set_xticklabels(labels)

plt.show()
```
